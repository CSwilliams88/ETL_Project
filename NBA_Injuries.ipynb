```python

import pandas as pd

# import the dataframe from the csv
data = "Resources/injuries_2010-2020.csv"
injuries_df = pd.read_csv(data, encoding="utf-8")
injuries_df

```

```python
# remove unneeded info from data frame
new_injuries_df = injuries_df[["Date", "Team"]]
new_injuries_df
```

```python
# create columns for month and date with datetime
new_injuries_df["Date"] = pd.to_datetime(new_injuries_df["Date"])
new_injuries_df["Year"] = new_injuries_df["Date"].dt.year
new_injuries_df["Date"] = pd.to_datetime(new_injuries_df["Date"])
new_injuries_df["Month"] = new_injuries_df["Date"].dt.month

#drop nulll values
clean_injury_df = new_injuries_df.dropna(how="any")
clean_injury_df
```

```python

# determine the season for injuries occuring in the first half of the season (fall)
fall_df = clean_injury_df.loc[clean_injury_df["Month"]>6]
fall_df["first"]=fall_df["Year"]
fall_df["last"]=fall_df["Year"]-1999
fall_df['first']=fall_df['first'].apply(str)
fall_df['last']=fall_df['last'].apply(str)

# assign the correct name of the season
fall_df["SEASON"]=fall_df["first"]+"-"+fall_df["last"]
fall_df


```

```python
# determine the season for injuries occuring in the second half of the season (spring)
spring_df = clean_injury_df.loc[clean_injury_df["Month"]<6]
spring_df["first"]=spring_df["Year"]-1
spring_df["last"]=spring_df["Year"]-2000
spring_df['first']=spring_df['first'].apply(str)
spring_df['last']=spring_df['last'].apply(str)

# assign the correct name of the season
spring_df["SEASON"]=spring_df["first"]+"-"+spring_df["last"]
spring_df

```

```python
# combine the fall and spring halves of the season
final_injuries_df = pd.concat([fall_df,spring_df])
final_injuries_df.head(40)
```

```python
# expand the nicknames to the full names of teams to correspond with data from another source

new_names=[]
for name in final_injuries_df["Team"]:
    if name == "Hawks": new_name = "Atlanta Hawks"
    elif name == "Celtics": new_name = "Boston Celtics"
    elif name == "Nets": new_name = "Brooklyn Nets"
    elif name == "Hornets": new_name = "Charlotte Hornets"
    elif name == "Bulls": new_name = " Chicago Bulls"
    elif name == "Cavs": new_name = "Cleveland Cavaliers"
    elif name == "Mavs": new_name = "Dallas Mavericks"
    elif name == "Nuggets": new_name = "Denver Nuggets"    
    elif name == "Pistons": new_name = "Detroit Pistons"
    elif name == "Warriors": new_name = "Golden State Warriors"
    elif name == "Rockets": new_name = "Houston Rockets"
    elif name == "Pacers": new_name = "Indiana Pacers"
    elif name == "Clippers": new_name = "LA Clippers"
    elif name == "Lakers": new_name = "Los Angeles Lakers"
    elif name == "Grizzlies": new_name = "Memphis Grizzlies"
    elif name == "Heat": new_name = "Miami Heat"
    elif name == "Bucks": new_name = "Milwaukee Bucks"
    elif name == "Timberwolves": new_name = "Minnesota Timberwolves"
    elif name == "Pelicans": new_name = "New Orleans Pelicans"
    elif name == "Knicks": new_name = "New York Knicks"    
    elif name == "Thunder": new_name = "Oklahoma City Thunder"
    elif name == "Magic": new_name = "Orlando Magic"
    elif name == "76ers": new_name = "Philadelphia 76ers"
    elif name == "Suns": new_name = "Phoenix Suns"
    elif name == "Blazers": new_name = "Portland Trailblazers"
    elif name == "Kings": new_name = "Sacramento Kings"
    elif name == "Spurs": new_name = "San Antonio Spurs"
    elif name == "Raptors": new_name = "Toronto Raptors"
    elif name == "Jazz": new_name = "Utah Jazz"
    elif name == "Wizards": new_name = "Washington Wizards"
    new_names.append(new_name)
new_names
```

```python
#add the expanded names to the dataframe
final_injuries_df["Full Name"]=new_names

# create the key by combining the full team name and the season
final_injuries_df["Team Season"]=final_injuries_df["SEASON"] + " " + final_injuries_df["Full Name"]
final_injuries_df.head(40)
```

```python
#group by the team season key
grouped_injuries_df = final_injuries_df.groupby(['Team Season'])

#count the total injuries per team by season
injuries = grouped_injuries_df['Team Season'].count()
```

```python
#create final dataframe
injury_summary_df = pd.DataFrame({"Number of Injuries": injuries,})
injury_summary_df
```

```python
#export the final table to csv
injury_summary_df.to_csv("Resources/injury_counts.csv", index=True, header=True)
```
